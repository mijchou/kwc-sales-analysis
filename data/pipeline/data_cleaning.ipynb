```python
import os
import pandas as pd
from dbfread import DBF

# Set pandas display option to show up to 50 columns
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)


# Get the current directory where the Jupyter notebook is located
current_directory = #os.getcwd()

# Get a list of all .dbf files in the directory
dbf_files = sorted([file for file in os.listdir(current_directory) if file.endswith('.DBF')])

print(f'Total file number: {len(dbf_files)}')
print(dbf_files)
```

### All DBS files & columns

```python
for dbf_file in dbf_files:
    # Read the .dbf file
    dbf_path = os.path.join(current_directory, dbf_file)
    table = DBF(dbf_path)
    
    # Convert DBF to pandas DataFrame
    df = pd.DataFrame(iter(table))
    
    # Print the name of the file
    print(f"File Name: {dbf_file}")
    
    # Print the column names
    print("Column Names:", df.columns.tolist())
    
    # Print the first 10 rows of the DataFrame
    # print("First 10 Rows:\n", df.head(10))
    # print("\n" + "="*40 + "\n")
```

### .DBF to .CSV

```python
## folder
CSV_folder = 'CSV/'

## files locations
STOCK_DBF = 'STOCK.DBF'
SORDERHD_DBF = 'SORDERHD.DBF'
ACCOUNTS_DBF = 'ACCOUNTS.DBF'
STTRANS_DBF = 'STTRANS.DBF'

dbf_files = [STOCK_DBF, SORDERHD_DBF, ACCOUNTS_DBF, STTRANS_DBF]
```

```python
# Read convert to pandas df
for i in dbf_files:
    table = DBF(i)
    df = pd.DataFrame(iter(table))
    df.to_csv(CSV_folder + i.split('.')[0] + '.CSV', index=False)
    # print(CSV_folder + i.split('.')[0] + '.CSV')
```
